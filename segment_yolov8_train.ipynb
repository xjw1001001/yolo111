```python
def format_boxes(boxes, resize = False, x_start = 0, y_start = 0):
    results = []
    for i in range(len(boxes.xyxy)):
        # convert from xyxy to xywh format
        x1, y1, x2, y2 = boxes.xyxy[i]
        
        if resize == True:
            # Resize the coordinates from 512 to 768
            x1, x2 = [x * (768 / 512) for x in [x1, x2]]
            y1, y2 = [y * (768 / 512) for y in [y1, y2]]

            # Add the start position
            x1 += x_start
            x2 += x_start
            y1 += y_start
            y2 += y_start
        
        width = x2 - x1
        height = y2 - y1
        
        # get the class and confidence score
        cls = int(boxes.cls[i].item())
        conf = float(boxes.conf[i].item())

        # add the result to the list
        results.append({
            "class": cls,
            "bbox": [float(x1), float(y1), float(width), float(height)],
            "prob": conf,
        })

    return results

def plot_bboxes(image_path, bboxes):
    # Load the image
    image = cv2.imread(image_path)

    # OpenCV uses BGR, convert it to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    for bbox_dict in bboxes:
        # retrieve bbox, class, and probability
        bbox = bbox_dict['bbox']
        cls = bbox_dict['class']
        prob = bbox_dict['prob']

        # Bounding box coordinates
        x, y, w, h = bbox
        x = int(x)
        y = int(y)
        w = int(w)
        h = int(h)

        # Drawing bounding box
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # Writing class and confidence
        cv2.putText(image, f'{cls}: {prob:.2f}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    plt.imshow(image)
    plt.show()

import re

def parse_filename(filename):
    # Regular expression pattern to extract name, xstart, and ystart
    pattern = r'^(.+)_(\d+)_(\d+)\.png$'

    match = re.match(pattern, filename)

    if match:
        name, x_start, y_start = match.groups()
        return name, int(x_start), int(y_start)

    else:
        raise ValueError(f"Filename {filename} doesn't match the expected format.")

# Test the function
filename = "T39_Rd2_FAP_Arm_A_10-2-2-2_73000x_1536_2688.png"
name, x_start, y_start = parse_filename(filename)
img = "/home/xionjing/CyroEM_segment/Cocodataset/train/images/T39_Rd2_FAP_Arm_A_16-5-3-5_73000x_0_3072.png"

# Run batched inference on a list of images
results = model([img, img], stream=True)  # return a generator of Results objects
# Process results generator
for result in results:
    boxes = result.boxes  # Boxes object for bbox outputs
    
bbox = format_boxes(boxes)
plot_bboxes(img, bbox)
```

```python
# Convert all of the boxes back the the original ones.
import os
import cv2
import json
from PIL import Image
from glob import glob
import re


model_path = "/home/xionjing/br-bards-digitalpathology/CyroEM_segmentation/YoloV8/HPVseg/yolov8x-12/weights/best.pt"
 # figure folder path    
__TEST_PATH = "/home/xionjing/CyroEM_segment/Cocodataset/train/images"
model = YOLO(model_path)


# Get all the images
image_paths = glob(os.path.join(__TEST_PATH, "*.png"))
results_dict = {}

for image_path in image_paths:
    # Extract the x and y start positions from the filename
    filename = os.path.basename(image_path)
    name, y_start, x_start = parse_filename(filename)

    # Load image
    img = Image.open(image_path)

    # Run inference
    result = model(img)

    # Process the result
    boxes = result[0].boxes
    
    # Iterate over each detected bounding box
    bbox = format_boxes(boxes, resize = True, x_start = x_start, y_start = y_start)

    # Add to results dictionary
    if name not in results_dict.keys():
        results_dict[name] = bbox
    else:
        results_dict[name] = results_dict[name] + bbox

# Save to json file
with open("results_train.json", "w") as f:
    json.dump(results_dict, f)
```
